```python
import matplotlib.pyplot as plt
from qiskit.visualization import plot_bloch_vector

# Coordenadas Bloch de los estados
bloch_states = {
    "H": [0, 0, 1],
    "V": [0, 0, -1],
    "D": [1, 0, 0],
    "A": [-1, 0, 0],
    "R": [0, 1, 0],
    "L": [0, -1, 0],
}

fig, axes = plt.subplots(1, 6, figsize=(18, 4), subplot_kw={"projection": "3d"})

for ax, (name, vec) in zip(axes, bloch_states.items()):
    plot_bloch_vector(vec, title=name, ax=ax)

plt.tight_layout()
plt.show()
```
