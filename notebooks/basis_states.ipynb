```python

from py_pol.jones_vector import Jones_vector
from py_pol.utils import degrees

# Horizontal
JH = Jones_vector('H')
JH.linear_light(azimuth=0*degrees)
```

```python
# Vertical
JV = Jones_vector('V')
JV.linear_light(azimuth=90*degrees)
```

```python
# Diagonal
JD = Jones_vector('D')
JD.linear_light(azimuth=45*degrees)
```

```python
# Antidiagonal
JA = Jones_vector('A')
JA.linear_light(azimuth=-45*degrees)
```

```python
# Right circular
JR = Jones_vector('R')
JR.circular_light(kind='r')
```

```python
# Left circular
JL = Jones_vector('L')
JL.circular_light(kind='l')
```
