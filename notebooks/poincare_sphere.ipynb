```python
from py_pol import degrees
from py_pol.stokes import Stokes

# Base states
S_h = Stokes('H')
S_h.linear_light(azimuth=0 * degrees)

S_v = Stokes('V')
S_v.linear_light(azimuth=90 * degrees)

S_d = Stokes('D')
S_d.linear_light(azimuth=45 * degrees)

S_a = Stokes('A')
S_a.linear_light(azimuth=-45 * degrees)

S_r = Stokes('R')
S_r.circular_light(kind='right')

S_l = Stokes('L')
S_l.circular_light(kind='left')

# Group for plotting
todos = S_h + S_v + S_d + S_a + S_r + S_l

#poincaré sphere
fig = todos.draw_poincare(kind="scatter", figsize=(8, 8))
fig.show()
```
